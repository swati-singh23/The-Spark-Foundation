# K-Means Clustering

```python
#We'll use the iris dataset to perform K-Means clustering and determine the optimal number of clusters.
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Function to find the optimum number of clusters using the elbow method
def find_optimal_clusters(data, max_clusters=10):
    wcss = []
    for i in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    return wcss
```

```python
# Plot the elbow method to find the optimal number of clusters
def plot_elbow(wcss):
    plt.plot(range(1, len(wcss) + 1), wcss)
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')  # Within cluster sum of squares
    plt.show()
```

```python
# Using the elbow method to find the optimal number of clusters
wcss = find_optimal_clusters(iris_df)
plot_elbow(wcss)
```

```python
# Explanation: The optimum number of clusters is where the elbow occurs, indicating a significant decrease in WCSS.

# Visualizing the clusters using a different approach
def visualize_clusters(data, clusters):
    plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis')
    plt.title('Visualizing Clusters')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.colorbar(label='Cluster')
    plt.show()
```

```python

# Applying K-Means with the optimal number of clusters and visualizing the clusters
optimal_clusters = 3  # Chosen based on the elbow method
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(iris_df)
visualize_clusters(iris_df.values[:, :2], clusters)
```
